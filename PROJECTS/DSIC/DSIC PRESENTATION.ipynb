# Introduction

## What is Git? 

“Git is an open-source version control system created by Linus Torvalds in 2005.  

Specifically, Git is a distributed version control system, which means that the entire   
codebase and history is available on every developer’s computer, which allows for easy   
branching and merging.   

According to a Stack Overflow developer survey, over 87% of developers use Git.”  

Source: https://kinsta.com/knowledgebase/what-is-github/

## Why is Version Control Important?

The basic documentation of all code work, managing and tracking changes,  
allows developers and teams to maintain code integrity and the ability  
to track and revert to changes.
